## **Data Cleaning**

```python
train.info()
```

```python
train.isnull().sum()
```

```python
train.drop(["Cabin"], axis=1, inplace=True)
```

```python
train["Age"] = train["Age"].fillna(train["Age"].mean())
```

```python
train["Embarked"].value_counts()
```

```python
train["Embarked"]=train["Embarked"].fillna("S")
```

```python
train.isnull().sum()
```
